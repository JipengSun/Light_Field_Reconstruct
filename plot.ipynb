```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-white')

sparsity_x = [64.0,37.03703703703704,23.323615160349853,15.625,10.973936899862826,8.0]
visibility_y = [0.984,0.9444444444444444,0.8717201166180758,0.828125,0.7489711934156379,0.616]
plt.grid()
plt.title('The Relationship Between Visibility and Sparsity')
plt.xlabel('Sparsity (Volume of Sample / Number of Sample)')
plt.ylabel('Visibility (Visible Points Number / Overall Points Number)')
plt.plot(sparsity_x,visibility_y,'r*')
plt.plot(sparsity_x,visibility_y)
for i, j in zip(sparsity_x, visibility_y):
   plt.text(i+2, j, '({}, {})'.format(np.round(i,2), np.round(j,2)))
```
